```python
from aac_metrics import evaluate
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/result/clip_easy_5.csv')
df['1'] = df['1'].transform(lambda x: [x])
prediction = df['0'].tolist()
references = df['1'].tolist()
ce_corpus_scores, _ = evaluate(prediction, references)
print(ce_corpus_scores)
```

```python
df = pd.read_csv('/content/drive/MyDrive/result/clip_hard_5.csv')
df['0'] = df['0'].transform(lambda x: x.replace("\n", " "))
df['1'] = df['1'].transform(lambda x: [x])
prediction = df['0'].tolist()
references = df['1'].tolist()
ch_corpus_scores, _ = evaluate(prediction, references)
print(ch_corpus_scores)
```

```python
df = pd.read_csv('/content/drive/MyDrive/result/blip_easy_5.csv')
df['1'] = df['1'].transform(lambda x: [x])
prediction = df['0'].tolist()
references = df['1'].tolist()
be_corpus_scores, _ = evaluate(prediction, references)
print(be_corpus_scores)
```

```python
df = pd.read_csv('/content/drive/MyDrive/result/blip_hard_5.csv')
df['0'] = df['0'].transform(lambda x: x.replace("\n", " "))
df['1'] = df['1'].transform(lambda x: [x])
prediction = df['0'].tolist()
references = df['1'].tolist()
bh_corpus_scores, _ = evaluate(prediction, references)
print(bh_corpus_scores)
```

```python
metrics = [["mertic", "ce", "ch", "be", "bh"]]
for x in ce_corpus_scores.keys():
  metric = [x]
  metric.append(ce_corpus_scores[x].item())
  metric.append(ch_corpus_scores[x].item())
  metric.append(be_corpus_scores[x].item())
  metric.append(bh_corpus_scores[x].item())
  metrics.append(metric)
display(metrics)

df = pd.DataFrame(metrics)
new_header = df.iloc[0]
df = df[1:]
df.columns = new_header
display(df)
```
